### WITHOUT FINE TUNING

```python
import matplotlib.pyplot as plt
import pandas as pd

# Given updated data for plotting
gad_data = {
    "Metric": ["Precision", "Recall", "F1-score"],
    "BioBERT": [52.84, 50.11, 33.23],
    "Gemma2b": [24.03, 40.58, 30.00],
    "Gemma 7b": [47.10, 49.80, 35.62],
    "Llama2 7b": [27.47, 49.37, 34.20]
}

euadr_data = {
    "Metric": ["Precision", "Recall", "F1-score"],
    "BioBERT": [40.73, 42.57, 29.22],
    "Gemma2b": [34.65, 40.07, 37.16],
    "Gemma 7b": [36.86, 49.80, 42.37],
    "Llama2 7b": [36.56, 48.28, 41.61]
}

# Convert dictionaries to DataFrames and add a 'Dataset' column
gad_df = pd.DataFrame(gad_data)
euadr_df = pd.DataFrame(euadr_data)
gad_df['Dataset'] = 'GAD'
euadr_df['Dataset'] = 'EU-ADR'

# Combine the two DataFrames into one for plotting
combined_df = pd.concat([gad_df, euadr_df], ignore_index=True)
```

```python
combined_df
```

```python

# Plot setup
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
metrics = ["Precision", "Recall", "F1-score"]

# Plotting each metric
for i, metric in enumerate(metrics):
    subset = combined_df[combined_df['Metric'] == metric]
    ax = subset.plot(
        kind='bar',
        x='Dataset',
        y=["BioBERT", "Gemma2b", "Gemma 7b", "Llama2 7b"],
        ax=axes[i],
        title=f"{metric} by Dataset"
    )
    axes[i].set_ylabel(f"{metric} Score")
    axes[i].set_xlabel("Dataset")
    axes[i].set_ylim([0, 100])  # Adjust Y-axis to range from 0 to 100
    axes[i].grid(True)
    axes[i].legend(title='Model', loc='upper left')

plt.tight_layout()
plt.show()
```
