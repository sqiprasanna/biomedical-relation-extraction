## FEATURE NAMES
1. Article identifier (PMID)
2. Title and Abstract
3. Entity or term number (for this record)
4. Type of entity mention (CHEMICAL, GENE-Y, GENE-N)*
5. Chemical-Protein relation (CPR) group
6. Start character offset of the entity mention
7. End character offset of the entity mention
8. Text string of the entity mention
9. CPR group
10. Evaluation type (Y: group evaluated, N: group not evaluated  extra annotation).
11. CHEMPROT relation (CPR)
12. interactor argument 1 (Arg1: followed by the interactor term identifier)
13. interactor argument 2 (Arg2: followed by the interactor term identifier)





Group Eval. CHEMPROT relations belonging to this group
1. CPR:1 N PART_OF
2. CPR:2 N REGULATOR|DIRECT_REGULATOR|INDIRECT_REGULATOR
3. CPR:3 Y UPREGULATOR|ACTIVATOR|INDIRECT_UPREGULATOR
4. CPR:4 Y DOWNREGULATOR|INHIBITOR|INDIRECT_DOWNREGULATOR
5. CPR:5 Y AGONIST|AGONIST-ACTIVATOR|AGONIST-INHIBITOR
6. CPR:6 Y ANTAGONIST
7. CPR:7 N MODULATOR|MODULATOR-ACTIVATOR|MODULATOR-INHIBITOR
8. CPR:8 N COFACTOR
9. CPR:9 Y SUBSTRATE|PRODUCT_OF|SUBSTRATE_PRODUCT_OF
10. CPR:10 N NOT
